#### **Question 2:** 
#### Data-visualization

Let's create plots to visualize the relationships between the features (sepal length, sepal width, petal length, petal width).


```python
def plot_feature_pairs(data, feature_names, color_labels=None, title_prefix=''):
    """
    Helper function to create scatter plots for all possible pairs of features.
    
    Parameters:
    - data: DataFrame containing the features to be plotted.
    - feature_names: List of feature names to be used in plotting.
    - color_labels: Optional. Cluster or class labels to color the scatter plots.
    - title_prefix: Optional. Prefix for plot titles to distinguish between different sets of plots.
    """
    # Create a figure for the scatter plots
    plt.figure(figsize=(12, 10))
    
    # Counter for subplot index
    plot_number = 1
    
    # Loop through each pair of features
    for i in range(len(feature_names)):
        for j in range(i + 1, len(feature_names)):
            plt.subplot(len(feature_names)-1, len(feature_names)-1, plot_number)
            
            # Scatter plot colored by labels if provided
            if color_labels is not None:
                plt.scatter(data[feature_names[i]], data[feature_names[j]], 
                            c=color_labels, cmap='viridis', alpha=0.7)
            else:
                plt.scatter(data[feature_names[i]], data[feature_names[j]], alpha=0.7)
            
            plt.xlabel(feature_names[i])
            plt.ylabel(feature_names[j])
            plt.title(f'{title_prefix}{feature_names[i]} vs {feature_names[j]}')
            
            # Increment the plot number
            plot_number += 1

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Show the plot
    plt.show()


# Get feature names
feature_names = iris_df.columns

# Use the helper function to plot scatter plots without coloring by cluster labels
plot_feature_pairs(iris_df, feature_names, title_prefix='Original Data: ')
```

#### K-means clustering 

Apply the K-Means clustering algorithm to the Iris dataset. Choose the value 3 for the number of clusters (`k=3`) and fit the model. Assign cluster labels to the original data and add them as a new column in the DataFrame.

```python
# Your answer...
k=3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(iris_df[iris_df.columns[:-2]])
iris_df['Cluster'] = kmeans.labels_
clustered_iris_data_2 = iris_df


# Use the helper function to plot scatter plots, colored by cluster labels
plot_feature_pairs(clustered_iris_data_2, feature_names, color_labels=clustered_iris_data_2['Cluster'], title_prefix='Clustered Data: ')
```

```python
# Load iris dataset
iris_data = load_iris()

# Create a DataFrame from the dataset
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)

# Add species column
iris_df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)

# Set random seed
random_seed = 42

# Standardizing features
scaler = StandardScaler()

# Ensure 'species' exists and drop it only for the features to be standardized
iris_features = iris_df.drop("species", axis=1)

# Standardize the features
iris_standardized = scaler.fit_transform(iris_features)

# Fit KMeans
kmeans = KMeans(n_clusters=3, random_state=random_seed)
iris_df['Cluster'] = kmeans.fit_predict(iris_standardized)

# Create a copy of the DataFrame with clusters
clustered_iris_data_1 = iris_df.copy()

# Get feature names (ensure only feature columns are included)
feature_names = iris_df.columns[:-2]  # Exclude 'species' and 'Cluster' columns



# Plot clustered data
plot_feature_pairs(clustered_iris_data_1, feature_names, 
                  color_labels=clustered_iris_data_1['Cluster'], 
                  title_prefix='Clustered Data: ')
```

```python
# Compare the two clustering results
print("Comparison of two clustering approaches:")
comparison = pd.crosstab(clustered_iris_data_1['Cluster'], 
                        clustered_iris_data_2['Cluster'],
                        margins=True)
print("\nCross-tabulation of Cluster Assignments:")
print(comparison)

# Compare cluster means
print("\nFirst approach cluster means:")
print(clustered_iris_data_1.groupby('Cluster')[feature_names].mean())

print("\nSecond approach cluster means:")
print(clustered_iris_data_2.groupby('Cluster')[feature_names].mean())

# Visualize the differences
plt.figure(figsize=(15, 5))

# Plot first approach
plt.subplot(1, 2, 1)
plt.scatter(clustered_iris_data_1[feature_names[0]], 
           clustered_iris_data_1[feature_names[1]], 
           c=clustered_iris_data_1['Cluster'],
           cmap='viridis')
plt.title('First Approach Clustering')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])

# Plot second approach
plt.subplot(1, 2, 2)
plt.scatter(clustered_iris_data_2[feature_names[0]], 
           clustered_iris_data_2[feature_names[1]], 
           c=clustered_iris_data_2['Cluster'],
           cmap='viridis')
plt.title('Second Approach Clustering')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])

plt.tight_layout()
plt.show()
```

#### **Question 5:** 
#### Bootstrapping 

Implement bootstrapping on the mean of Petal Width. Generate 10000 bootstrap samples, calculate the mean for each sample, and compute a 90% confidence interval.

```python
# Your answer here...
petal_width = iris_df['petal width (cm)']
n_samples = 10000
bootstrap_means = [np.mean(np.random.choice(petal_width, size=len(petal_width), replace=True)) for _ in range(n_samples)]

mean_petal_width = np.mean(petal_width)

np.random.seed(123)

lower_bound = np.percentile(bootstrap_means, 5)
upper_bound = np.percentile(bootstrap_means, 95)



# Display the result
print(f"Mean of Petal Width: {mean_petal_width}")
print(f"90% Confidence Interval of Mean Petal Width: ({lower_bound}, {upper_bound})")
```

**Question:**
- Why do we use bootstrapping in this context? What does it help us understand about the mean?

>  based on text book Data Science - A First Introduction page 332: We use bootstapping because we have just one smaple from our population (we don't have access to the full population itself). As such, we need to approximate what the sampling distribution would look like for the sample and then use that approximation to report the uncertainty of the sample point estimate. We are estimating the distribution of the mean by repeatedly sampling from the data with replacement, to understand the sampling variability of the mean of the sepal_length feature from the Iris datas.


**Question:**
- What is the purpose of calculating the confidence interval from the bootstrap samples? How does it help us interpret the variability of the estimate?

> based on text book Data Science - A First Introduction page 332: confidence interval - the range of plausible values for the population parameter It helps us assess the uncertainty and variability of the estimate by giving a range within which the true population parameter is likely to fall The confidence interval of (5.72 cm, 5.98 cm) for the mean sepal length suggests that the true mean is likely to fall within this range, which tells us how much uncertainty there is around the sample mean of 5.84 cm.
